```python
def sanitize_input(input_string, chars_to_remove):
    # Create a translation table with None for characters to remove
    translation_table = str.maketrans('', '', chars_to_remove)
    # Translate the input string using the translation table
    sanitized_string = input_string.translate(translation_table)
    return sanitized_string

# Example usage
input_string = "Hello, World! Welcome to Python."
chars_to_remove = ",!."
sanitized_string = sanitize_input(input_string, chars_to_remove)
print(sanitized_string)  # Output: "Hello World Welcome to Python"
```
